```python
x = np.linspace(0,np.pi*2,30)
y = np.sin(x)
y1 = np.cos(x)
y2 = np.tan(x)
y3=x
figure,axis = plt.subplots(2,2)
axis[0,0].plot(x,y)
axis[0,0].set_title("Sine Function")
axis[0,0].set_xlabel("angle")
axis[0,0].set_ylabel("sine")

axis[0,1].plot(x,y1)
axis[0,1].set_title("cosine Function")
axis[0,1].set_xlabel("angle")
axis[0,1].set_ylabel("cosine")

axis[1,0].plot(x,y2)
axis[1,0].set_xlabel("angle")
axis[1,0].set_title("Tangent Function")
axis[1,0].set_ylabel("Tangent")

axis[1,1].plot(x,y3)
axis[1,1].set_xlabel("angle")
axis[1,1].set_ylabel("Cos")
axis[1,1].set_title("Cos function")
```

```python
x = [i for i in range(100)]
y = [i**2 for i in x ]
z = [i**3 for i in x]
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax1.plot(x,y)
ax1.set_title("Quadratic")
ax2 = fig.add_subplot(1,2,2)
ax2.plot(x,y,z)
ax2.set_title("Cubic")
plt.show()
```
